## Embedding single-cell data into biologically inspired 2D space: enterocytes and intestinal villi

```python
import matplotlib.pyplot as plt
import pylab
import numpy as np
import scipy as sp
import pandas as pd
import skimage as ski
from skimage import io, color, filters
import seaborn as sns
from A_functions import transp_imshow
entero_path = "three_villi_1.png"
sns.set_context('talk')
from matplotlib import rc
rc('text', usetex=True)
```

### 1) Preparation of villi-shaped mask and grid

```python
# import the enterocyte image to create a mask
villi = np.invert(io.imread(entero_path)[:,:,3])
io.imshow(villi)
sns.despine()
np.shape(villi)
```

```python
# start with x and y vectors, then anumerate all potential grid points
x = np.linspace(0, np.shape(villi)[1], 50, endpoint=False)
y = np.linspace(0, np.shape(villi)[0], 50, endpoint=False)
all_points = [(xi, yi) for xi in x for yi in y]
```

```python
# now, which of these points lie within the villi?
# as an approximation, we will use the rounded versions of x and y coordinates as indices
# and then read out whether the mask value at that pixel is true or false
all_points_rounded = np.round(all_points)
```

```python
# turn villus image into mask
val = filters.threshold_otsu(villi)
mask = villi < val
```

```python
# in a new list, collect the points for which the mask says True
villus_points = [p for p in all_points_rounded if mask[int(p[1]), int(p[0])] == True]
len(villus_points)
```

```python
# add noise to x and y of villus points, to make final visualizations less griddy
noi = 1.5
villus_points_jitter = [(x[0]+noi*np.random.rand(), x[1]+noi*np.random.rand()) for x in villus_points]
# now, sort the point by their y values
villus_points_jitter = sorted(villus_points_jitter, key=lambda x: x[1], reverse=True)
# extract the x and y with jitter
x_j = [x[0] for x in villus_points_jitter]
y_j = [x[1] for x in villus_points_jitter]
```

### 2) Import single-cell data, sample the required amount and prepare bin-averages for task-mean plots

```python
#villus_obs = pd.read_csv('/Users/lbuchauer/owncube/F_datavis/bioembed/sc_villus_obs.csv', index_col=0)
#villus_obs = pd.read_csv('/Users/lisa/Nextcloud/F_datavis/bioembed/sc_villus_obs.csv', index_col=0)
villus_obs = pd.read_csv('sc_villus_obs.csv', index_col=0)
villus_data = pd.read_csv('sc_villus_count_data.csv', index_col=0)
# preprocess data, for scatterplots lognormalized data is required
villus_data_norm = villus_data.div(villus_data.sum(axis=1), axis=0).multiply(1e4)
villus_data_log1p = villus_data.apply(lambda x: np.log(x+1))
```

```python
# add eta and zone value to count data and log'ed data
# Moor et al. zones are only required to remove crypt cells from the dataset
villus_data['eta'] = villus_obs['eta']
villus_data['zone'] = villus_obs['zone']
villus_data_log1p['eta'] = villus_obs['eta']
villus_data_log1p['zone'] = villus_obs['zone']
```

```python
# divide data into 10 bins along eta
bins = np.linspace(0, 1, 11)
villus_data['binned'] = pd.cut(villus_data['eta'], bins=bins)
villus_data_log1p['binned'] = pd.cut(villus_data_log1p['eta'], bins=bins)
```

```python
# remove crypt as we are plotting villus cells here
villus_data = villus_data[villus_data['zone'] != 'Crypt']
villus_data_log1p = villus_data_log1p[villus_data_log1p['zone'] != 'Crypt']
```

```python
# first three genes in the lists below will be plotted individually, but all will be
# included in task average
apolipo_genes = ['Apoa1', 'Apoa4', 'Apobec1', 'Apob', 'Npc1l1']
carb_genes = ['Slc2a2', 'Slc2a5', 'Slc5a1']
amino_genes = ['Slc7a7', 'Slc7a8', 'Slc7a9']
reg_genes = ['Reg1', 'Reg3b', 'Reg3g', 'Nlrp6', 'Lypd8', 'Il18', 'Ccl25']
combi_genes = apolipo_genes + carb_genes + amino_genes + reg_genes
genes_eta = combi_genes + ['eta', 'binned']
# subset count and log data
select_data = villus_data[genes_eta]
select_data_log1p = villus_data_log1p[genes_eta]
```

```python
# sample the dataframe according to the number of points to be drawn
# if there are enough cells, do not replace; if there are more points than
# availablce cells, sample with replacement
# only necessary for logged data as this is plotted in scatter
if len(villus_points) <= len(villus_data):
    data_sample = select_data_log1p.sample(n=len(villus_points), replace=False)
else:
    data_sample = select_data_log1p.sample(n=len(villus_points), replace=True)
```

```python
print(len(villus_data))
print(len(villus_points))
```

```python
# sort the dataframe by eta
data_sample = data_sample.sort_values('eta')
```

```python
# Make average profiles for the gene sets, happens on count data
zone_data = select_data.groupby('binned').mean()
```

```python
# groupwise mean along zones
apolipo_mean = zone_data[apolipo_genes].mean(axis=1).values
carb_mean = zone_data[carb_genes].mean(axis=1).values
amino_mean = zone_data[amino_genes].mean(axis=1).values
reg_mean = zone_data[reg_genes].mean(axis=1).values
```

```python
# interpolate data to a vector that is as long as the villus image is high
from scipy.interpolate import interp1d
from scipy.signal import savgol_filter as sgf
def interp_mean_data(old_mean, x=np.linspace(0, 1, 10), typ='mean'):
    y = sgf(old_mean, window_length=3, polyorder=1)
    f = interp1d(x, y, kind='slinear')
    if typ=='mean':
        xnew = np.arange(np.shape(villi)[0])/(np.shape(villi)[0]-1)
    elif typ=='scatter':
        xnew = np.arange(np.max(y_j))/np.max(y_j)
    interp_mean = f(xnew)
    return interp_mean
```

```python
apolipo_interp = interp_mean_data(apolipo_mean)
carb_interp = interp_mean_data(carb_mean)
amino_interp = interp_mean_data(amino_mean)
reg_interp = interp_mean_data(reg_mean)
```

```python
# define a function which returns the eta value at which a gene profile peaks
def max_eta(gene, typ='scatter'):
    # first, smooth and interpolate the binned gene values
    gene_mean = zone_data[[gene]].mean(axis=1).values
    gene_interp = interp_mean_data(gene_mean, typ=typ)
    max_eta = np.argmax(gene_interp)
    return max_eta
```

```python
# fill a grid with these column vectors
apolipo_map = np.flipud(np.vstack([apolipo_interp for x in range(np.shape(villi)[1])]).T)
carb_map = np.flipud(np.vstack([carb_interp for x in range(np.shape(villi)[1])]).T)
amino_map = np.flipud(np.vstack([amino_interp for x in range(np.shape(villi)[1])]).T)
reg_map = np.flipud(np.vstack([reg_interp for x in range(np.shape(villi)[1])]).T)
```

```python
# test plot
plt.imshow(reg_map*mask, cmap='magma', vmin=0)
transp_imshow(villi, 0, 1, cmap='Greys_r')
plt.axis('off')
plt.show()
```

### 4) Plot it!

```python
# define general parameters
s = 14
alpha = 1

# define task order and colors
tasks = [apolipo_genes, carb_genes, amino_genes, reg_genes]
task_grads = [apolipo_map, carb_map, amino_map, reg_map]
labels = ['Lipid\nuptake', 'Carbohydrate\nuptake', 'Amino acid\nuptake', 'Bacterial\ndefense']
#colors = sns.husl_palette(7, l=0.75, h=0.30, s=1)[3:]
colors = sns.husl_palette(4, l=0.75, h=0.55, s=1)[::-1]
base = np.array([20, 20, 20])/255  # base color for starting colormaps

# initialize figure
fig, axes = plt.subplots(4,4, figsize=(14, 11), sharey=False, sharex=True)

# for each task, plot a mean gradient plot and 3 scatter plots
# complete with label and colorbar
for i, genes in enumerate(tasks):
    
    # create palette from the first color
    pal = sns.blend_palette([base, colors[i]], n_colors=10, as_cmap=True)

    # in the first column, plot mean of task genes
    grad = axes[i][0].imshow(task_grads[i]*mask, cmap=pal, vmin=0)
    transp_imshow(villi, 0, 1, cmap='Greys_r', ax=axes[i][0])
    
    # based on this, plot colorbar onto a new axis to the left of the first
    cax = fig.add_axes([axes[i][0].get_position().x0 - 0.6*axes[i][0].get_position().width, 
                        axes[i][0].get_position().y0 + 0.*axes[i][0].get_position().height,
                        0.4*axes[i][0].get_position().width,
                        0.05])
    cbar = plt.colorbar(grad, ax=cax, aspect=5, orientation='horizontal', fraction=1, ticks=[0.01*np.max(task_grads[i]), 0.99*np.max(task_grads[i])])
    cbar.ax.set_xticklabels(['0', 'max'])
    cbar.outline.set_linewidth(0.5)
    cax.axis('off')
    
    # on the remaining three panels, plot the scatter plots of individual genes
    # scatter
    for j, ax in enumerate([axes[i][1], axes[i][2], axes[i][3]]):
        # first, get the maximum eta position, plot a horizontal line there
        meta = max_eta(tasks[i][j])
        ax.hlines(np.max(y_j)-meta, xmin=5, xmax=np.max(x_j)+5, color=colors[i], linewidth=15, alpha=0.4, zorder=0)
        col = data_sample[tasks[i][j]]
        ax.scatter(x_j, y_j, c=col, marker='o', s=s, cmap=pal, linewidth=0, alpha=alpha)
        ax.set_title(tasks[i][j], rotation='vertical', x=0.06, y=0.33)
        ax.invert_yaxis()
    
    # add a label describing the task above the colorbar
    xpos = cax.get_position().x0 + 0.5 * cax.get_position().width
    ypos = cax.get_position().y0 + 0.04
    fig.text(xpos, ypos, labels[i], bbox=dict(facecolor=colors[i], edgecolor='white', linewidth=2, alpha=0.4, boxstyle='round4'), fontsize=18, verticalalignment='center', horizontalalignment='center')

    
# finally, remove all axes and ensure equal aspect
for i, ax in enumerate(axes.flatten()):
    ax.set_aspect('equal')
    ax.axis('off')

# add an eta-axis arrow to the last last plot in the first row
axes[0][3].annotate('$\eta$', xy=(1, 0.03), xycoords='axes fraction', xytext=(0.98, 0.5), fontsize=16,
            arrowprops=dict(arrowstyle="<-", color=base))

axes[0][3].annotate(r'$\eta$ (max expression)', xy=(0.67, 0.93), xycoords='axes fraction', xytext=(0.37, 1.1), fontsize=16,
            arrowprops=dict(arrowstyle="->", color=base))

# add titles for all columns, have the same y position
ypos = 0.93
ft = 18  # fontsize

# 1) task label column
xpos1 = cax.get_position().x0 + 0.5 * cax.get_position().width
fig.text(xpos1, ypos, 'enterocyte\ntask', style='italic', bbox=dict(facecolor='white', edgecolor=None, alpha=0.5, linewidth=0, boxstyle='round'), fontsize=ft, verticalalignment='top', horizontalalignment='center')

# 2) mean expression column
xpos2 = axes[0][0].get_position().x0 + 0.55 * axes[0][0].get_position().width
fig.text(xpos2, ypos, 'mean expression of\ntask-related genes', style='italic', bbox=dict(facecolor='white', edgecolor=None, alpha=0.5, linewidth=0, boxstyle='round'), fontsize=ft, verticalalignment='top', horizontalalignment='center')

# 3) scatter columns
xpos3 = axes[0][2].get_position().x0 + 0.5 * axes[0][2].get_position().width
fig.text(xpos3, ypos, 'expression of individual task-related genes in single cells', style='italic', bbox=dict(facecolor='white', edgecolor=None, alpha=0.5, linewidth=0, boxstyle='round'), fontsize=ft, verticalalignment='top', horizontalalignment='center')
                   
# 4) overall title
fig.text(0.5, 0.96, 'Spatial division of labor along villi of the small intestine', bbox=dict(facecolor='white', edgecolor=None, alpha=0.5, linewidth=0, boxstyle='round'), fontsize=22, verticalalignment='center', horizontalalignment='center')

# adjust whitespace
plt.subplots_adjust(hspace=0.07, wspace=0.07)

pylab.savefig('spatial_labor_div_intestine.pdf', bbox_inches='tight')
```
