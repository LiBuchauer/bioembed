```python
# we will need a function which makes points on a circle of given radius
def circle_points(r, n):
    t = np.linspace(0, 2*np.pi, n)
    x = r * np.cos(t)
    y = r * np.sin(t)
    return x, y
```

```python
# define some radii via interpolation
from scipy import interpolate
vill_x = [0, 0.5,  1.2, 2, 4, 6, 8, 9, 9.5, 10]
vill_r = [1, 0.75, 0.5, 0.5 ,0.5, 0.5, 0.5, 0.48, 0.4, 0.]
f = interpolate.interp1d(vill_x, vill_r, kind='quadratic')
z_val = np.arange(0, 10, 0.3)
radii = f(z_val)
plt.figure(figsize=(10,2))
plt.plot(z_val, radii, 'o')
```

```python
# make circle for each of the radii and at x,y, z to a list
xlist = []
ylist = []
zlist = []
pp1 = 50. # points per 0.1 radius

for i, r in enumerate(radii):
    x, y = circle_points(r, int(pp1*r))
    xlist += list(x)
    ylist += list(y)
    zlist += [z_val[i] for j in range(int(pp1*r))]
```

```python
# now turn one villus into 3 by adding the list with shifted x values
base_r = radii[0]
x_3 = xlist + list(np.array(xlist)+2*base_r) + list(np.array(xlist)+4*base_r)
y_3 = ylist + ylist + ylist
z_3 = zlist+ zlist + zlist
```

```python
# add some jitter to x, y and z  of villus points
noise = 0.1
x_3 = [x + noise * np.random.rand() for x in x_3]
y_3 = [x + noise * np.random.rand() for x in y_3]
z_3 = [x + 2*noise * np.random.rand() for x in z_3]

```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(5, 85)
ax.scatter(x_3, y_3, z_3, s=10, marker='o', alpha=1, c=np.random.rand(len(x_3)))
ax.axis('off')
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_trisurf(x_3, y_3, z_3, linewidth=0)
fig.colorbar(surf)

ax.view_init(5, 85)

ax.axis('off')
fig.tight_layout()
```
