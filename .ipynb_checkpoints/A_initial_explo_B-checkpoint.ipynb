### Initial exploration for embedding sc data in villus shapes

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import pandas as pd
import skimage as ski
from skimage import io, color, filters
import seaborn as sns
entero_path = "/Users/lisa/Nextcloud/F_datavis/bioembed/three_villi_1.png"
#entero_path = "/Users/lbuchauer/owncube/F_datavis/bioembed/three_villi_1.png"
sns.set_context('talk')
```

```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
# import the enterocyte image to create a mask
villi = np.invert(io.imread(entero_path)[:,:,3])
io.imshow(villi)
sns.despine()
np.shape(villi)
```

```python
# start with x and y vectors, then anumerate all potential grid points
x = np.linspace(0, np.shape(villi)[1], 50, endpoint=False)
y = np.linspace(0, np.shape(villi)[0], 50, endpoint=False)
all_points = [(xi, yi) for xi in x for yi in y]
```

```python
# now, which of these points lie within the villi?
# as an approximation, we will use the rounded versions of x and y coordinates as indices
# and then read out whether the mask value at that pixel is true or false
all_points_rounded = np.round(all_points)
```

```python
# turn villus image into mask
val = filters.threshold_otsu(villi)
mask = villi < val
```

```python
# in a new list, collect the points for which the mask says True
villus_points = [p for p in all_points_rounded if mask[int(p[1]), int(p[0])] == True]
len(villus_points)
```

```python
# plot a scatter of villus points only
plt.figure(figsize=(10, 10*(np.shape(villi)[0]/np.shape(villi)[1])))
plt.scatter([x[0] for x in villus_points], [x[1] for x in villus_points], marker='h')
plt.gca().invert_yaxis()
```

### Import actual data

```python
#villus_obs = pd.read_csv('/Users/lbuchauer/owncube/F_datavis/bioembed/sc_villus_obs.csv', index_col=0)
villus_obs = pd.read_csv('/Users/lisa/Nextcloud/F_datavis/bioembed/sc_villus_obs.csv', index_col=0)
```

```python
#villus_data = pd.read_csv('/Users/lbuchauer/owncube/F_datavis/bioembed/sc_villus_count_data.csv', index_col=0)
villus_data = pd.read_csv('/Users/lisa/Nextcloud/F_datavis/bioembed/sc_villus_count_data.csv', index_col=0)
```

```python
# provide option to normalise
#villus_data = villus_data.div(villus_data.sum(axis=1), axis=0).multiply(1e4)
```

```python
# add eta and zone value to data
villus_data['eta'] = villus_obs['eta']
villus_data['zone'] = villus_obs['zone']
```

```python
# remove crypt as we are plotting villus cells here
villus_data = villus_data[villus_data['zone'] != 'Crypt']
```

```python
# we don't actually need so many genes, let's just keep a few for now
test_genes = ['Lypd8',
         'Nlrp6',
         'Il18',
         'Fth1',
         'Gstm3',
         'Dpp4',
         'Slc5a1',
         'Cd74',
         'Cdh1',
         'Apoa4',
         'Fos',
         'Egfr',
         'Nt5e',
         'Slc28a2',
         'Actb',
         'Gapdh',
         'Tuba1b',
         'Reg3b',
         'Ada']
apolipo_genes = ['Apoa1', 'Apoa4', 'Apob']
carb_genes = ['Slc2a2', 'Slc2a5', 'Slc5a1']
amino_genes = ['Slc7a7', 'Slc7a8', 'Slc7a9']
reg_genes = ['Reg1', 'Reg3b', 'Reg3g']
combi_genes = apolipo_genes + carb_genes + amino_genes + reg_genes
genes_eta = combi_genes + ['Fos', 'eta', 'zone']
select_data = villus_data[genes_eta]
```

```python
# sample the dataframe according to the number of points to be drawn
# if there are enough cells, do not replace; if there are more points than
# availablce cells, sample with replacement
if len(villus_points) <= len(villus_data):
    data_sample = select_data.sample(n=len(villus_points), replace=False)
else:
    data_sample = select_data.sample(n=len(villus_points), replace=True)
```

```python
# sort the dataframe by eta
data_sample = data_sample.sort_values('eta')
```

### Plot villus with real data

```python
# first, add noise to x and y of villus points
noi = 1.5
villus_points_jitter = [(x[0]+noi*np.random.rand(), x[1]+noi*np.random.rand()) for x in villus_points]
# now, sort the point by their y values
villus_points_jitter = sorted(villus_points_jitter, key=lambda x: x[1], reverse=True)
# extract the x and y with jitter
x_j = [x[0] for x in villus_points_jitter]
y_j = [x[1] for x in villus_points_jitter]
```

```python
# plot zones
col = data_sample['zone']
plt.figure(figsize=(4, 5*(np.shape(villi)[0]/np.shape(villi)[1])))
sns.scatterplot(x_j, y_j, hue=col, marker='o', s=20, linewidth=0)
plt.gca().invert_yaxis()
plt.axis('off')
plt.legend(bbox_to_anchor=(1.05, 0.5), loc=2)
plt.title('Zone')
```

```python
# now, plot some genes on a grid
fig, axes = plt.subplots(3,4, figsize=(15,11))
for i, ax in enumerate(axes.flatten()):
    col = np.log(data_sample[combi_genes[i]]+1)
    sns.scatterplot(x_j, y_j, hue=col, marker='o', s=15, ax=ax, legend=False, palette='viridis', linewidth=0)
    ax.invert_yaxis()
    ax.axis('off')
    ax.set_title(combi_genes[i])
```
